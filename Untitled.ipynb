```python
df['Genre_Principal'] = df['Genre_Tags'].str.split(' / ').str[0]

print(" Genre principal extrait :")
print(df[['Anime', 'Genre_Tags', 'Genre_Principal']].head(10))
```

```python
df['Ecart'] = df['Note_Meilleur_Ep'] - df['Note_Pire_Ep']
```

```python
df['Regularite'] = 10 - df['Ecart']
```

```python
df["Score_Qualite"] = (df["Note_Globale"] * 0.5) + (df["Regularite"] * 0.5)
```

```python
import numpy as np
df["Score_Final"] = df["Score_Qualite"] + (np.log1p(df["Nb_Episodes"]) * 0.1)
```

```python
display(df[["Anime", "Note_Globale", "Regularite", "Score_Qualite"]].sort_values(by="Score_Qualite", ascending=False).head(10))
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import KMeans
X = df[['Score_Qualite', 'Regularite', 'Note_Globale', 'Score_Final']]
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Segment'] = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
features = df[['Note_Globale', 'Regularite', 'Nb_Episodes','Score_Final']].values
similarite = cosine_similarity(features)
anime_id = 10
scores_similarite = list(enumerate(similarite[anime_id]))
scores_tries = sorted(scores_similarite, key=lambda x: x[1], reverse=True)[1:6]
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import KMeans

X = df[['Score_Qualite', 'Regularite', 'Note_Globale', 'Score_Final']]

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=42)
df['Segment'] = kmeans.fit_predict(X_scaled)
```

```python
df['Segment'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    data=df,
    x='Score_Qualite',
    y='Score_Final',
    hue='Segment',
    palette='tab10'
)
plt.title("Visualisation des clusters K-Means")
plt.savefig("Visualisation_des_clusters_KMeans.pdf", bbox_inches='tight')
plt.show()
```

```python
import pandas as pd

centers = kmeans.cluster_centers_
centers_df = pd.DataFrame(centers, columns=X.columns)
centers_df
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, df['Segment'])
score
```

```python
def recommander(anime, n=10):
    seg = df.loc[df['Anime'] == anime, 'Segment'].values[0]
    return df[df['Segment'] == seg].sort_values('Score_Final', ascending=False).head(n)
```

```python
recommander("Naruto")
```
